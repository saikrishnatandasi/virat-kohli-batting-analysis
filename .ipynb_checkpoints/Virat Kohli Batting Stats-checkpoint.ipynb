### ODI Stats Scraping

```python
odi_request = requests.get(odi_url,headers = {"User-agent":user_agents["Google Chrome (Windows)"]})
```

```python
print(odi_request.status_code)
print(odi_request.reason)
```

```python
odi_soup = BeautifulSoup(odi_request.content,"html.parser")
```

```python
odi_tables = odi_soup.find_all("table", class_= "engineTable")
```

```python
print(odi_tables)
```

```python
print(type(odi_tables))
```

```python
print(len(odi_tables))
```

```python
odi_table = odi_tables[3]
```

```python
odi_headers = [i.text for i in odi_table.find_all("th")]
print(odi_headers)
```

```python
odi_headers = [ i for i in odi_headers if i != ""]
print(odi_headers)
```

```python
odi_headers.append("Scorecard")
print(odi_headers)
```

```python
odi_values = []
for i in odi_table.find_all("tr")[1:]:
#     print(i.text)
    odi_values.append([j.text for j in i.find_all("td")if j.text != ""])
        
print(odi_values)
```

```python
odi_df = pd.DataFrame(odi_values, columns = odi_headers)
odi_df
```

```python
odi_df.to_csv("virat_odi_stats.csv", index = False)
```
